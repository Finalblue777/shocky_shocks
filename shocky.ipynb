```python
import numpy as np
import math as mp
```

```python
# Some parameter initialization. 
sigma = 10
beta = 0.9
W = 1
N = np.linspace(8,10,3)
A = np.array([[1,1,1],[1,1,0.9],[1,0.9,1],[0.9,1,1], [1,1,1]])
```

```python

tau= -1/np.log(beta) - N
print(tau)
tau[tau<0] = 0  
tau = np.ceil(tau)
tprod =  beta**tau * (tau + N)
tprod
```

```python
tau
```

```python
P10 = (np.sum(1/(A*tprod)**(1-sigma),axis =1))**(1/(1-sigma))
P1 = (np.sum(1/(A*tprod)**(1-1.1),axis =1))**(1/(1-1.1))
P0 = (np.sum(1/(A*tprod)**(1-0.5),axis =1))**(1/(1-0.5))
P5 = (np.sum(1/(A*tprod)**(1-5),axis =1))**(1/(1-5))
```

```python
import matplotlib.pyplot as plt
plt.plot( (1/P10)*P10[0], label='sigma = 10')
plt.plot( (1/P1)*P1[0], label='sigma = 1.1')
plt.plot( (1/P0)*P0[0], label='sigma = 0.5')
plt.plot( (1/P5)*P5[0], label='sigma = 5')

plt.legend()
plt.title(f'Comparison between different values of sigma')
plt.xlabel('t')
plt.ylabel('Normalized Output')
plt.show()
```
