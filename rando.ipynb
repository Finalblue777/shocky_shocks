```python
import numpy as np
import math as mp
```

```python
# Some parameter initialization. 
sigma = 10
beta = 0.9
W = 1
N = np.linspace(8,10,3)
A = np.array([[1,1,1],[1,1,0.9],[1,0.9,1],[0.9,1,1], [1,1,1]])
```

```python

tau= -1/np.log(beta) - N
print(tau)
tau[tau<0] = 0  
tau = np.ceil(tau)
tprod =  beta**tau * (tau + N)
tau
```

```python
P = (np.sum(1/(A*tprod)**(1-sigma),axis =1))**(1/(1-sigma))
```

```python
import matplotlib.pyplot as plt
plt.plot( (1/P), label='sigma = 10')


plt.legend()
plt.title(f'Comparison between different values of sigma')
plt.xlabel('t')
plt.ylabel('Normalized Output')
plt.show()
```
