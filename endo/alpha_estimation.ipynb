```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skewnorm

# Parameters: skewness, location, scale
skew = -6       # Negative = left skew
loc = 1.0       # Center the bulk of the data around 1
scale = 0.5     # Controls the spread

# Generate data
N = skewnorm.rvs(a=skew, loc=loc, scale=scale, size=10000) * 4
# Plot histogram
plt.hist(N, bins=30, color='slateblue', edgecolor='black', density=True)
plt.title(f'Skew-Normal Distribution (a={skew}, loc={loc}, scale={scale})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()

# Print fraction of negative values
print(f"Fraction negative: {np.mean(N < 0):.3f}")
```
